## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

_______________________________________________________________________________________________________________________________

Primary Keys are unique keys assigned to each entry in a table. Foreign Keys are column values that are used to align with Primary Keys of another table. Joining two tables with Primary Keys and Foreign Keys can add additional rows and/or columns of data to the original table.

## Question 3

_______________________________________________________________________________________________________________________________

One-to-One: Each entry in a table is linked to only one entry in another table.

One-to-Many: Each entry in a table is linked to multiple entries in another tables.

Many-to-Many: Multiple entries in one table are linked to multiple entries in another table.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

INNER JOIN: Joins two tables where a key in the first table matches a key in the second table and returns values of matching keys.

OUTER JOIN: Joins two tables where there are matching and unmatching rows and returns values based on whether FULL, LEFT, or RIGHT OUTER JOIN is used.

LEFT JOIN: Joins two tables where the second table has keys matching in the first table, returning the full first table along with matches from the second table. NULL values are added where keys do not match.

RIGHT JOIN: The reverse of LEFT JOIN. SQLite does not recognize RIGHT JOIN.

FULL JOIN: Joins two tables, entirely, returning values for matching and unmatching keys and adding NULL values to unmatching keys. SQLite does not recognize FULL JOIN.
